```python
import tmdbsimple as tmdb
import requests
import pandas as pd
from config import tmdb_apikey
from config import omdb_apikey
from pprint import pprint
import numpy as np
```

```python
# Import original csv
movieDF = pd.read_csv("data/new_movieDF.csv")
movieDF1 = movieDF.drop(columns=  ['Unnamed: 0'])
movieDF1.head(50)
testDF = movieDF1.head(5)
testDF
```

```python
# URL Configs (TMDB)
# URL to search by movie title sample: https://api.themoviedb.org/3/search/movie?api_key={api_key}&query=Jack+Reacher
# URL to search by movie ID sample: https://api.themoviedb.org/3/movie/{movie_id}?api_key=<<api_key>>&language=en-US
# URL to search by keyword sample: https://api.themoviedb.org/3/search/multi?api_key=[apikey]&language=en-US&query=bob%20the%20hbuilder

baseURL = "https://api.themoviedb.org/3/movie/"
movie_id = testDF['TMDB ID'].unique()



# test queryURL 
queryURL = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_apikey}&language=en-US"
#testURL = f"https://api.themoviedb.org/3/search/multi?api_key={tmdb_apikey}&language=en-US&query={keyword}"
movieJSON = requests.get(queryURL).json()
#movieJSON
```

```python
# TMDB API
# Create a response list to store column data

# Title = movieJSON['title']
# Release Date = movieJSON['release_date']
# Genre = movieJSON['genres'][0]['name']
# Budget = movieJSON['budget']
# Revenue = movieJSON['revenue']
# Country = movieJSON['production_countries'][0]['name']

title = []
#release_date = []
genre_list = []
movie_budget = []
movie_revenue = []
country = []

print("Beginning Data Retrieval:\n"
     "-----------------------------------")

# Set counters
index_counter = 0
set_counter = 1

movies = testDF['TMDB ID']

# Make a request for each movie ID using a for loop
for i, movieID in enumerate (movies, start = 1):

    try:
        response = requests.get(queryURL).json()
        title.append(response['title'])
        genre_list.append(response['genres'][0]['name'])
        movie_budget.append(response['budget'])
        movie_revenue.append(response['revenue'])
        country.append(response['production_countries'][0]['name'])
        
#         if index_counter > 49:
#             index_counter = 0
#             set_counter = set_counter + 1
#         else:
#             index_counter = index_counter + 1
#         print(f"Processing Record {index_counter} of Set {set_counter} | {movieID}")
        
    except(KeyError, IndexError):
        
        print("Movie not found. Skipping...")
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
testDF1 = pd.DataFrame({
    'TItle' : title, 
    'Genre' : genre_list, 
    'Budget' : movie_budget, 
    'Revenue' : movie_revenue, 
    'Production Country' : country
})
testDF1
```

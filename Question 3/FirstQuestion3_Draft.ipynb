```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
from config import apiKey
import time
import tmdbsimple as tmdb
```

```python
KaggleDF = pd.read_csv('../Data/cleaned_moviesDF1.csv', index_col=False)
KaggleDF
```

```python
movie_titles = KaggleDF['Title'].tolist()
years = KaggleDF['Year'].tolist()

TestList = KaggleDF['Year'].tolist()
stringList = [str(x) for x in years]
tmdb.REQUESTS_TIMEOUT = 5 
tmdb.API_KEY = apiKey
```

```python
url = "https://api.themoviedb.org/3/search/movie?api_key="
queryURL = url + apiKey + "&query="

#f"https://api.themoviedb.org/3/search/movie?api_key={apiKey}m&query={movie_title}&year={year}"

#set up lists for dataFrame
movieTitle = []
yearsAPI =[]
popularity = []
genres = []
IDlist = []
languageList = []
voteAverage = []
voteCount = []
moviesFound = []
moviesNotFound = []
record = 1
setC = 1


for movie, year in zip(movie_titles,stringList):
    movieResponse = requests.get(queryURL+movie +"&year="+year).json()
    print(f"Processing Record {record} of set {setC} //")
    
    if record < 60:
        try:
            moviesFound.append(movieResponse['results'][0]['original_title'])
            yearsAPI.append(movieResponse['results'][0]['release_date'])
            IDlist.append(movieResponse['results'][0]['id'])
            languageList.append(movieResponse['results'][0]['original_language'])
            popularity.append(movieResponse['results'][0]['popularity'])
            genres.append(movieResponse['results'][0]['genre_ids'])
            voteAverage.append(movieResponse['results'][0]['vote_average'])
            voteCount.append(movieResponse['results'][0]['vote_count'])
            print(f"Movie '{movie}' of '{year}' found, counted")
        except:
            print(f"Movie '{movie}' of '{year}' not found, skipped")
            moviesNotFound.append(movie)
        
        record+=1
    else:
        time.sleep(5)
        record = 1
        setC+=1


# print(f"Data retrieval finished\n{len(moviesFound)} was the number of movies found found while {len(moviesNotFound)} did not have any data")
```

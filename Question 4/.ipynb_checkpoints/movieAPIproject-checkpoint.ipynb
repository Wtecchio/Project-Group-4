```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from api_key import api_key
dataToLoad = "../data/productionCountriesWithoutNaNs.csv"
kagMovDF = pd.read_csv(dataToLoad)
movie_id = ""
url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}"
```

```python
kagMovDF["Correct ID"].count()
kagIMDBLan = []

kagIMDBLan = kagMovDF[kagMovDF["Correct ID"].notna()]

#kagMovDF.dropna(axis=0, subset = ["Correct ID"], inplace = True)

kagIMDBLan
```

```python
prodCountry = []
idSave = []

for index, movie in kagIMDBLan.iterrows():
    movie_id = int(kagIMDBLan["Correct ID"][index])
    url1 = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}"
    movie_data = requests.get(url1).json()
    try:
        prodCountry.append(movie_data['production_countries'][0]['name'])
        idSave.append(movie_id)
        print(f"Trying {movie_id}")
    except:    
        prodCountry.append(np.NaN)
        idSave.append(movie_id)
        print("Could not find Production Country")
        
    
    
print(prodCountry)
    
#requests.get(url + movie).json()
#resultDF = pd.DataFrame(movie_data['results'])
#resultDF['release_date'] = pd.to_datetime(resultDF['release_date']).dt.year
#resultDrop = resultDF.drop(resultDF.loc[resultDF["release_date"] < 2012].index)
#resultDrop
```

```python
prodCountryDict = {'Production Country':prodCountry,'Correct ID': idSave}
prodCountryDF = pd.DataFrame(prodCountryDict)
prodCountryMerge = pd.merge(prodCountryDF, kagIMDBLan, on = 'Correct ID')
prodCountryMerge
```

```python
prodCountryMerge.to_csv(r'productionCountriesWithNaNs.csv')

```

```python
prodCountryMergeClean = prodCountryMerge[prodCountryMerge["Production Country"].notna()]
len(prodCountryMergeClean)
```
